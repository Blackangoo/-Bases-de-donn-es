Tableau de données, partie 1

```python
######### PERSONNE #############


def generate_data_personne(personne_ID,personne_usernames,personne_noms,personne_prenoms,personne_passwords,personne_emails):

    return {
        "ID": personne_ID,
        "Nom d'utilisateur": personne_usernames,
        "Nom": personne_noms,
        "Prénom": personne_prenoms,
        "Mot de passe": personne_passwords,
        "Email": personne_emails }


######### LABEL #############


def generate_data_label(label_ID,swiss_labels):

    return {
        "ID": label_ID,
        "Label": swiss_labels
    }


######### ARTISTE #############


def generate_data_artiste(artiste_biographies_clean,artiste_label,total_entries):
    personne_df = pd.read_csv('personne.csv')
    unique_df = personne_df.drop_duplicates(subset=['Nom', 'Prénom'])
    selected_df = unique_df.sample(n=total_entries//15, random_state=42)
    artiste_ID = selected_df['ID']
    return {
        "ID": artiste_ID,
        "Biographie": artiste_biographies_clean,
        "ID_Label": artiste_label
    }


######### EVENEMENT #############


def generate_data_evenement(evenement_id,evenement_lieu,evenement_date,event_liens):
    
    return {
        "ID": evenement_id,
        "Lieu": evenement_lieu,
        "Date": evenement_date,
        "Lien": event_liens
    }


######### ABONNEMENT #############


def generate_data_abo(abo_id, abo_prix,abo_type,abo_description):

    return {
        "ID": abo_id,
        "Prix": abo_prix,
        "Type": abo_type,
        "Abonnement": abo_description
    }

def generate_data_user(user_id,user_annee,user_canton,user_abo):
    artiste_df = pd.read_csv('artiste.csv')
    artiste_id = artiste_df['ID'].values
    for k in range(1,1+total_entries):
        if k not in artiste_id:
            user_id.append(k)
            x = random.randint(1940, 2010)
            user_annee.append(x)
            user_canton.append(random.choices(swiss_cantons, weightsCantons)[0])
            generate_abo(x,user_abo)
    return {
        "ID": user_id,
        "Année de naissance": user_annee,
        "Canton": user_canton,
        "Abonnement": user_abo
    }
    

    ######### PRODUIT DERIVES #############


def generate_data_goodies(goodies_id,goodies_type,goodies_prix,goodies_links):

    return {
        "ID": goodies_id,
        "Goodie": goodies_type,
        "Prix": goodies_prix,
        "Lien de vente": goodies_links
    }


    ######### VENTES #############


def generate_data_ventes(ventes_id_artiste,ventes_id_produit):
    artiste_df = pd.read_csv('artiste.csv')
    artiste_id = artiste_df['ID'].values
    for artiste in artiste_id:
        produits=random.randint(1,29)
        while produits<29:
            ventes_id_artiste.append(artiste)
            ventes_id_produit.append(produits)
            y=random.randint(1,10)
            produits=produits+y
    return {
        "ID_artiste": ventes_id_artiste,
        "ID_produit_derives": ventes_id_produit
    }


    ######### ARTISTE FAVORI #############


def generate_data_favart(favart_artiste_id, favart_user_id, favart_note):
    artiste_df = pd.read_csv('artiste.csv')
    artiste_id = artiste_df['ID'].values
    user_df = pd.read_csv('utilisateur.csv')
    user_id = user_df['ID'].values
    for user in user_id:
        for artiste in artiste_id:
            x=random.randint(1,100)
            if x<6:
                favart_artiste_id.append(artiste)
                favart_user_id.append(user)
                favart_note.append(x)
    return {
        "ID_utilisateur": favart_user_id,
        "ID_artiste": favart_artiste_id,
        "Note": favart_note
    }




```

```python
######### VIDEO CLIP #############


def generate_data_videoclip(videoclip_id, list_animation,videoclip_duree):

    return {
        "ID": videoclip_id,
        "Animation": list_animation,
        "Durée": videoclip_duree }


######### ALBUM #############


def generate_data_album(album_ids, album_name, album_date):

    return {
        "ID": album_ids,
        "Titre de l'album": album_name,
        "Date de sortie": album_date
    }


######### PLAYLIST #############


def generate_data_playlist(playlist_id,playlist_nom):

    return {
        "ID": playlist_id,
        "Nom": playlist_nom
    }


######### CONTENU AUDIO #############


def generate_data_contenu_audio(song_ids,song_name,song_duree,song_date,song_parole_clean, videoclip_id,album_song_ID):
    return {
        "ID": song_ids,
        "Titre": song_name,
        "Durée": song_duree,
        "Date de sortie": song_date,
        "Paroles": song_parole_clean,
        "ID video_clip": videoclip_id,
        "ID album": album_song_ID
    }


######### APPARTENANCE #############


def generate_data_appartenance(appartenance_id_contenu, appartenance_id_playlist):
    audio_df = pd.read_csv('contenu_audio.csv')
    audio_id = audio_df['ID'].values
    playlist_df = pd.read_csv('playlist.csv')
    playlist_id = playlist_df['ID'].values
    for playlist in playlist_id:
        nbr_audio = len(audio_id)
        audio = random.randint(1, nbr_audio)
        while audio < nbr_audio:
            appartenance_id_playlist.append(playlist)
            appartenance_id_contenu.append(audio_id[audio - 1])  # Utiliser l'ID réel du contenu audio
            x = nbr_audio // 10
            y = random.randint(1, x)
            audio += y

    return {
        "ID_contenu_audio": appartenance_id_contenu,
        "ID_playlist": appartenance_id_playlist
    }


######### ECOUTE #############


def generate_data_ecoute(ecoute_user_ids, ecoute_audio_ids, note, nombre_ecoute, last_listening,ecoute_total):
    
    audio_df = pd.read_csv('contenu_audio.csv')
    audio_id = audio_df['ID'].values
    user_df = pd.read_csv('utilisateur.csv')
    user_id = user_df['ID'].values
    for user in user_id:
        nbr_audio = len(audio_id)
        audio = random.randint(1, nbr_audio)
        while audio < nbr_audio:
            ecoute_total.append(generate_listen_time())
            ecoute_user_ids.append(user)
            ecoute_audio_ids.append(audio_id[audio - 1])  # Utiliser l'ID réel du contenu audio
            x=random.randint(1,5)
            note.append(x)
            nombre_ecoute.append(200*x+random.randint(-150,150))
            date = random_date(start_date, end_date).strftime('%Y-%m-%d')
            last_listening.append(date)
            x = nbr_audio // 10
            y = random.randint(1, x)
            audio += y

    return {
        "ID_contenu_audio": ecoute_audio_ids,
        "ID_utilisateur": ecoute_user_ids,
        "Note": note,
        "Nombre d'écoute": nombre_ecoute,
        "Date dernière écoute": last_listening,
        "Temps_ecoute" : ecoute_total
    }


######### LANGAGE #############


def generate_data_langage(langue_id, langue):

    return {
        "ID": langue_id,
        "Langue": langue
    }


######### LANGUE UTILISATEUR #############


def generate_data_langue_user(langue_user_ids, langue_langage_ids):
    user_df = pd.read_csv('utilisateur.csv')
    user_id = user_df['ID'].values
    for user in user_id:
        langue_user_ids.append(user)
        langue_langage_ids.append(random.randint(1,4))
    return {
        "ID_user": langue_user_ids,
        "ID_Langage": langue_langage_ids
    }


######### PARLE #############


def generate_data_parle(song_ids, new_langage):

    return {
        "ID_contenu": song_ids,
        "ID_Langage": new_langage
    }


   ######### GENRE #############


def generate_data_genre(genres_music,genres_id):
    return {
        "ID": genres_id,
        "Genre": genres_music
    }
    

    ######### DECRIT #############


def generate_data_goodies(decrits_audio_ids,decrits_genre_ids,genre_weights):
    audio_df = pd.read_csv('contenu_audio.csv')
    audio_id = audio_df['ID'].values
    for audio in audio_id:
        decrits_audio_ids.append(audio)
        x = random.choices(range(1, 26), genre_weights)[0]
        decrits_genre_ids.append(x)
    return {
        "ID_contenu_audio": decrits_audio_ids,
        "ID_genre": decrits_genre_ids
    }


    ######### CREDITS #############


def generate_data_credits(credit_ids, list_artiste, credit_song_ids):
    return {
        "ID_credit": credit_ids,
        "ID_artiste": list_artiste,
        "ID_contenu" : credit_song_ids
    }


    ######### TYPE ARTISTE #############


def generate_data_type_art(type_ids,instrument_players ):

    return {
        "ID": type_ids,
        "Type_artiste" : instrument_players
    }


    ######### MENTIONNE #############


def generate_data_mentionne(mentionne_credits_id,mentionne_type_id):
    credit_df = pd.read_csv('credit.csv')
    credits_id = credit_df['ID_credit'].values
    for credits in credits_id:
        mentionne_credits_id.append(credits)
        mentionne_type_id.append(random.randint(1,10))

    return {
        "ID_credits": mentionne_credits_id,
        "ID_type_artiste": mentionne_type_id
    }


    ######### PARTICIPATION #############


def generate_data_participation(part_evenement_id, part_artiste_id):
    artiste_df = pd.read_csv('artiste.csv')
    artiste_id = artiste_df['ID'].values
    event_df = pd.read_csv('evenement.csv')
    event_id = event_df['ID'].values
    for event in event_id:
        part_evenement_id.append(event)
        x = random.randint(0, len(artiste_id) - 1)
        part_artiste_id.append(artiste_id[x])
    return {
        "ID_artiste": part_artiste_id,
        "ID_evenement": part_evenement_id
    }




```
